## Instalar Azure OpenAI Service .NET SDK

[Azure.AI.OpenAI NuGet package](https://www.nuget.org/packages/Azure.AI.OpenAI/)

````polyglot-notebook
#r "nuget: Azure.AI.OpenAI, 1.0.0-beta.5"
````

````polyglot-notebook
using System;
using Azure.AI.OpenAI;
````

## Configurar Azure OpenAI client

````polyglot-notebook
var AOAI_ENDPOINT = Environment.GetEnvironmentVariable("AOAI_ENDPOINT");
var AOAI_KEY = Environment.GetEnvironmentVariable("AOAI_KEY");
var AOAI_DEPLOYMENTID = Environment.GetEnvironmentVariable("AOAI_DEPLOYMENTID");
````

````polyglot-notebook
var endpoint = new Uri(AOAI_ENDPOINT);
var credentials = new Azure.AzureKeyCredential(AOAI_KEY);
var openAIClient = new OpenAIClient(endpoint, credentials);
````

## Definir indicaciones

````polyglot-notebook
var prompt = 
    """
    Convierte t√≠tulos de pel√≠culas en emojis.

    Volver al Futuro: üë®üë¥üöóüïí 
    Batman: ü§µü¶á 
    Transformers: üöóü§ñ 
    Star Wars:
    """;
````

## Configurar las opciones de las terminaciones

````polyglot-notebook
var completionOptions = new CompletionsOptions
{
    MaxTokens=60,
    Temperature=0.8f,
    FrequencyPenalty=0.0f,
    PresencePenalty=0.0f,
};

completionOptions.Prompts.Add(prompt);
````

## Generar terminaciones

````polyglot-notebook
Completions response = await openAIClient.GetCompletionsAsync(AOAI_DEPLOYMENTID,completionOptions);
````
